```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

THe hue parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

```python
data.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data)
```

```python
data.head()
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=data)

#jitter
```

```python
#swarmplot

sns.catplot(x='day',y='tip',kind='swarm',data=data)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='smoker',data=data)
```

```python
sns.swarmplot(x='day',y='tip',data=data)
```

```python
# box plot using seaborn
# used on top of numerical data
# univariate analysis

sns.boxplot(data['total_bill'])

```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='box',hue='sex',data=data)
```
