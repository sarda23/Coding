```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data=data)
```

Barplot and Countplot

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=data)
```

those bars(brownish) -- errors bars  -- confidence intervals to cal mean for the populations --- the value of mean for the population

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',estimator=np.var,data=data)
```

countplot is similar to histogram -- categorical data


```python
sns.catplot(x='sex',kind='count',hue='smoker',data=data)
```

Heatmaps


```python
flights=sns.load_dataset('flights')
flights.tail()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x,linewidths=0.5,annot=True,fmt='d') 
# d stands for integer
```

```python
sns.heatmap(x,linewidths=1,annot=True,fmt='d',cmap='summer') 
# d stands for integer
```

Distplots

1. histogram
2. PDF  kernel density enstimation
3. rugplot

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
