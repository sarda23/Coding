```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# pie charts    

```python
areas=['marketing','sales','development','hr','support']
budget=[2.5,4,10,1,6]
```

```python
plt.pie(budget,labels=areas,autopct="%0.1f%%")
plt.show()
```

```python
x=[23,45,12,49,97,32,11,44]
y=[45,34,67,87,55,22,89,90]

plt.scatter(x,y)
plt.xlabel('age')
plt.ylabel('sal')
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
```

```python
offline_price=[400,300,100,333,444,565,565,959]
chor_price=[100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id,online_price,color='black',marker='D',markersize=10,label="online")
plt.plot(product_id,offline_price,color='yellow',marker='+',markersize=10,label="offline prices")
plt.plot(product_id,chor_price,color='green',marker='+',markersize=10,label="chor prices")
plt.xlabel("product")
plt.ylabel("prices")
plt.title("product vs price comparison")
plt.legend()
plt.show()  # to remove the extra information
```

```python
year=[2012,2013,2014,2015,2016,2017,2018]

price=[65000,69000,75000,55000,52000,64000,81000]
```

```python
plt.bar(year,price,color='blue')
plt.xlabel('year')
plt.ylabel('price')
plt.title("year on year info")
plt.show()
```
